```python
import numpy as np

def get_top_tfidf_words(texts, top_n=15):
    vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
    tfidf_matrix = vectorizer.fit_transform(texts)
    feature_names = vectorizer.get_feature_names_out()

    mean_tfidf = np.asarray(tfidf_matrix.mean(axis=0)).ravel()
    top_indices = mean_tfidf.argsort()[::-1][:top_n]
    top_words = [(feature_names[i], mean_tfidf[i]) for i in top_indices]
    return top_words
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import umap.umap_ as umap
import matplotlib.pyplot as plt
import numpy as np
```

```python
vectorizer = TfidfVectorizer(stop_words='english', max_features=2000)
tfidf_matrix = vectorizer.fit_transform(df['Abstract'].dropna())
```

```python
n_clusters = 5
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
labels = kmeans.fit_predict(tfidf_matrix)
```

```python
df = df.loc[df['Abstract'].notna()]
```

```python
df['Cluster'] = labels
```

```python
reducer = umap.UMAP(random_state=42)
embedding = reducer.fit_transform(tfidf_matrix)
```

```python
plt.figure(figsize=(10, 7))
scatter = plt.scatter(embedding[:,0], embedding[:,1], c=labels, cmap='tab10', s=15)
plt.legend(*scatter.legend_elements(), title="Cluster")
plt.title('Visualisasi Clustering Abstrak dengan UMAP + K-Means')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.show()
```

```python
def top_keywords_per_cluster(tfidf_matrix, labels, feature_names, n_terms=10):
    df_cluster = pd.DataFrame(tfidf_matrix.toarray())
    df_cluster['cluster'] = labels
    for cluster_num in np.unique(labels):
        cluster_data = df_cluster[df_cluster['cluster'] == cluster_num].drop(columns='cluster')
        mean_tfidf = cluster_data.mean(axis=0)
        top_terms = mean_tfidf.sort_values(ascending=False).head(n_terms)
        print(f"\nTop kata cluster {cluster_num}:")
        for term in top_terms.index:
            print(f"{feature_names[term]}: {top_terms[term]:.4f}")

feature_names = vectorizer.get_feature_names_out()
top_keywords_per_cluster(tfidf_matrix, labels, feature_names)
```

```python
from collections import Counter
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))
```

```python
def get_word_counts(texts):
    all_words = " ".join(texts).lower().split()
    words = [word for word in all_words if word.isalpha() and word not in stop_words]
    return Counter(words).most_common(15)
```

```python
inflation_counts = get_word_counts(df_clean[df_clean['Label']=='Inflation']['Abstract'].dropna())
noninflation_counts = get_word_counts(df_clean[df_clean['Label']=='Non-Inflation']['Abstract'].dropna())
```
